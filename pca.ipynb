# Importing data and required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler as scale
```

```python
df=pd.read_csv("C:\\study\\my prep\\archive\\hotel_bookings.csv")
```

```python
df.info()
```

```python
# Count the number of NaNs each column has.
nans=pd.isnull(df).sum()
nans[nans>0]
```

```python
# Remove columns with more than a thousand missing values
df=df.drop("agent", 1)
df=df.drop("company", 1)
df=df.drop("reservation_status_date",1)
df=df.drop("country",1)
```

```python
#filling njll values of children
df['children']=df["children"].fillna(df['children'].mode()[0])
```

```python
#converting data type of children, adr to int
df["children"]=df["children"].astype(int)
df["adr"]=df["adr"].astype(int)
```

```python
#Count the column types
df.dtypes.value_counts()
```

```python
unique_counts = pd.DataFrame.from_records([(col, df[col].nunique()) for col in df.columns],
                          columns=['Column_Name', 'Num_Unique']).sort_values(by=['Num_Unique'])
```

```python
unique_counts
```

```python

categorical = ['hotel','is_canceled','arrival_date_year','arrival_date_week_number','arrival_date_month',
               'arrival_date_day_of_month','meal','market_segment','distribution_channel', 'is_repeated_guest','deposit_type ',
               'reserved_room_type','assigned_room_type','customer_type','reservation_status']
```

```python
for col in df.columns:
    if  col in categorical:
        df[col] = df[col].astype('category')
```

```python
df.info()
```

```python
#one hot encoding
for col in df.columns:
    if  col in categorical:
        data = pd.get_dummies(df)
```

```python
data.info()
```

```python
scaler = scale()
```

```python
scaler.fit(data[0:])
```

```python
df_pca = pd.DataFrame(scaler.transform(data[0:]))
df_pca
```

```python
pca =  PCA(n_components= 15).fit(df_pca)
df_pca = pd.DataFrame(pca.transform(df_pca))
pcs = ['pc'+str(i+1) for i in range(15)]
```

```python
df_pca.columns = pcs
df_pca.head()
```
